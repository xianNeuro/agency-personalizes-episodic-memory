```python
# 1. movie: compare between [different human ratings] and [gpt4 (prompt-v5) different trials]
import xianfunc as xf
prompt_v=5
story_ls = ['3','5','10','2','4','8']#['1020','1021','1027'] # ['4003','4004','4014'] # 
fromdir =  './6_chatgpt4agent/'#monthiversary_prolific/'
story_fls = xf.grab_all(fromdir,'movie*ptv'+str(prompt_v)+'_*xlsx')

outdir = './supplement/chatgpt4agent/'
xf.mkdir(outdir)

print(story_fls)
for story in story_fls:
    data = xf.read_data(fromdir+story,[1,2,3,4,5,6,7])
    data = xf.num_data_clean(data)
    isc_pw_keys = [i for i in data.keys() if 'pw' in i]
    isc_avg_keys = [i for i in data.keys() if 'avg' in i]
    
    xf.plot_figsize(10,6)
    xf.plot_hist(data[isc_pw_keys[0]],color='blue',add=True,bg_style='white',title=story.split('_')[0]+' Causal Centrality Pairwise-ISC: Human vs. GPT4',xylab=['isc_pearsonr','count'],legend=isc_pw_keys[0])
    xf.plot_hist(data[isc_pw_keys[1]],color='red',add=True,legend=isc_pw_keys[1])
    xf.plot_hist(data[isc_pw_keys[2]],color='green',legend=isc_pw_keys[2],add=True)
    outname = story.split('_')[0]+'_isc-pairwise_human_vs_gpt4-ptv'+str(prompt_v)
    xf.plot_save(outdir+outname,ftype='.svg')
    
```

```python
# 2. cyoa: compare between [different human ratings] and [gpt4 (prompt-v5) different trials]
import xianfunc as xf
story_ls =['1020','1021','1027','4003','4004','4014'] # 
fromdir =  './6_chatgpt4agent/'
story_fls = xf.grab_all(fromdir,'story*xlsx')

outdir = './6_chatgpt4agent/'
xf.mkdir(outdir)

print(story_fls)
for story in story_fls:
    data = xf.read_data(fromdir+story,[1,2,3,4,5,6,7])
    data = xf.num_data_clean(data)
    isc_pw_keys = [i for i in data.keys() if 'pw' in i]
    isc_avg_keys = [i for i in data.keys() if 'avg' in i]
    
    xf.plot_figsize(10,6)
    xf.plot_hist(data[isc_pw_keys[0]],color='blue',add=True,bg_style='white',title=story.split('_')[0]+' Causal Centrality Pairwise-ISC: Human vs. GPT4',xylab=['isc_pearsonr','count'],legend=isc_pw_keys[0])
    xf.plot_hist(data[isc_pw_keys[1]],color='red',add=True,legend=isc_pw_keys[1])
    xf.plot_hist(data[isc_pw_keys[2]],color='green',legend=isc_pw_keys[2],add=True)
    outname = story.split('_')[0]+'_isc-pairwise_human_vs_gpt4'
    xf.plot_save(outdir+outname,ftype='.svg')
    
    
```

```python
# 3. cyoa: compare betwween [human cuasal ratings vs. human memory] and [gpt4 causal ratings vs. human memory]
import xianfunc as xf
fromdir =  './6_chatgpt4agent/'
outdata = xf.num_data_clean(xf.read_data(fromdir+'compare_pred_recall.xlsx',[2,3]))
outdir = './6_chatgpt4agent/'

#adventure 3 stories
hum_causef = outdata['human_caus-rcl'][:15]
gpt_causef = outdata['gpt4_caus-rcl'][:15]

xf.plot_figsize(10,6)
xf.plot_hist(hum_causef,add=True,color='blue',bg_style='white',title='Adventure 3 stories: causal vs. recall',xylab=['causal-recall pearsonr','count'],legend='human')
xf.plot_hist(gpt_causef,color='green',legend='gpt4',outpath=outdir+'causal-recall_comparison_adv3stories',fig_format='.svg')

#romance 3 stories
hum_causef = outdata['human_caus-rcl'][15:]
gpt_causef = outdata['gpt4_caus-rcl'][15:]

xf.plot_figsize(10,6)
xf.plot_hist(hum_causef,add=True,color='blue',bg_style='white',title='Romance 3 stories: causal vs. recall',xylab=['causal-recall pearsonr','count'],legend='human')
xf.plot_hist(gpt_causef,color='green',legend='gpt4',outpath=outdir+'causal-recall_comparison_rom3stories',fig_format='.svg')
```
