# Scatter plots for idiosyncrasy, neighbor encoding, and semantic effect
1. romance story: recall diveregence from group vs. semantic centrality effect
2. romance story: recall diveregence from group vs. semantic centrality effect (exclude choice events)
3. romance story: neighbor encoding effect vs. semantic centrality effect 
4. adventure story: neighbor encoding effect vs. semantic centrality effect 
5. romance story: neighbor encoding effect vs. recall diveregence from group 
6. romance story: neighbor encoding effect vs. recall diveregence from group (without choice event) 

```python
# 1. romance story: recall diveregence from group vs. semantic centrality effect

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Load the Excel file
file_path = './2_scatter_plots/scatter_data.xlsx'
xls = pd.ExcelFile(file_path)

# Load the "romance" sheet into a DataFrame
df_romance = xls.parse('romance')

# Extract relevant columns
x = df_romance['recall-divergence-from-others']
y = df_romance['sem-ef']

# Calculate Pearson correlation coefficient and p-value
r_value, p_value = pearsonr(x, y)
r_text = f'r = {r_value:.3f}'
p_text = f'p = {p_value:.3f}' if p_value >= 0.001 else 'p < 0.001'
title_text = f'{r_text}  {p_text}'

# Create scatter plot with specified customizations, larger font sizes, and a title
plt.figure(figsize=(10, 7))
color = sns.color_palette("muted")[0]  # first color from seaborn muted palette
sns.scatterplot(x=x, y=y, color=color, s=200, marker='o', edgecolor='black')  # muted color, round dots with black outline and larger size
sns.regplot(x=x, y=y, scatter=False, color='black', line_kws={"linewidth":3}, ci=95, scatter_kws={'s':200, 'edgecolor':'black'})  # add thick black trend line with grey SE shade

# Set font sizes and adjust space from the labels
plt.xlabel('Recall divergence from group', fontsize=24, labelpad=20)
plt.ylabel('Semantic centrality\neffect on memory', fontsize=24, labelpad=20)
plt.xticks(fontsize=24)
plt.yticks(fontsize=24)
plt.title(title_text, fontsize=24)

# Adjust layout to provide more space for the y-axis label
plt.tight_layout(pad=2.5)

# Save the figure
png_path = './2_scatter_plots/scatter_rom_idio_vs_sem.png'
svg_path = './2_scatter_plots/scatter_rom_idio_vs_sem.svg'
plt.savefig(png_path, format='png')
plt.savefig(svg_path, format='svg')
plt.show()

```
