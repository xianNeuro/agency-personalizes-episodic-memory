```python
# 4. merged section: free-route subjects' recall temporal order

import xianfunc as xf

def intls2strls(int_ls):
    return [str(i) for i in int_ls]

def keep_merge_rcl(rcl,decipher=[obs,trn],mapf='monthy_map.xlsx'):
    out = rcl.copy()
    out_merge = []
    for i in range(len(rcl)):
        ls = rcl[i]
        temp = []
        for j in range(len(ls)):
            item = ls[j]
            inds = xf.get_indices(obs,str(item))
            trans = xf.apply_inds(inds,trn)
            out[i][j]=trans[0]
            merge_inds = xf.cyoa_get_merge(mapf,trans)
            if len(merge_inds)!=0:
                temp.append(item)
        out_merge.append(temp)
    rcl_merge = [','.join(intls2strls(i)) for i in out_merge if len(i)!=0]
    return rcl_merge

root = '../analysis/0_data/1_free/'
fls = xf.grab_all(root,'recall.xlsx')
exclude = ['sub2_4002_rate-recall.xlsx']
fls = [i for i in fls if i not in exclude]
len(fls)

outdir = '../'
xf.mkdir(outdir)

data = xf.init_dict(['subj','tv','tv_rate'])

for f in fls:
    subj = f.split('_rate')[0]
    print(subj)
    
    rcl = xf.read_data(root+f,[1])['recalled_events']
    rcl2 = [xf.process_recall_seg(i) for i in rcl]
    fname = f.split('rate-recall')[0]+'events.xlsx'
    eventinfo = xf.read_data(root+fname,[1,2])
    
    obs = eventinfo['event']
    trn = eventinfo['event_lab']
    rcl_merge = keep_merge_rcl(rcl2,decipher=[obs,trn])
    print(rcl_merge)
    
    mat = xf.recall_temp_mat(rcl_merge,tot)
    xf.plot_2dmat(mat,title=subj+'_Recall_Sequence',outpath=outdir)

    tv,tv_rate = xf.recall_temp_vio(rcl_merge)
    xf.dict_add_vals(data,['subj','tv','tv_rate'],[subj,tv,tv_rate])

xf.write_dict_anylen(data,'../analysis/10_temporal/merge/free.xlsx')
    
```
